```python
import plotly.graph_objs as go
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = -np.sin(x)  # Negative values to simulate below 0

# Create a figure
fig = go.Figure()

# Add the lines
fig.add_trace(go.Scatter(x=x, y=y1, mode='lines', name='Above 0'))
fig.add_trace(go.Scatter(x=x, y=y2, mode='lines', name='Below 0'))

# Fill between the lines with a color
fig.add_trace(go.Scatter(x=x, y=y1, fill='tozeroy', fillcolor='rgba(0, 255, 0, 0.3)', name='Fill'))

# Customize the layout
fig.update_layout(
    title='Fill Between Lines',
    xaxis=dict(title='X-axis'),
    yaxis=dict(title='Y-axis'),
    showlegend=True
)

# Show the plot
fig.show()
```

```python
import pandas as pd
from pydantic import BaseModel

class DataFrames(BaseModel):
    dataframe1: pd.DataFrame = None
    dataframe2: pd.DataFrame = None

    class Config:
        arbitrary_types_allowed = True
```

```python
from typing import Optional
from dataclasses import dataclass
from yahooquery import Ticker

def get_price(ticker):
    # logic to get price by looking at `name`.
    return Ticker(ticker).history(start=(datetime.today() - timedelta(days = 365*5)).strftime('%Y-%m-%d'), end=datetime.today().strftime('%Y-%m-%d'))

@dataclass
class Stock:
    symbol: str
    price: Optional[float] = None

    def __post_init__(self):
        if self.price is None:
            self.price = get_price(self.symbol)
```

```python
a = Stock(symbol='AAPL')
```

```python
from option import VanillaOption
from model import BlackScholesModel
from strategy import Strategy
from plot import plot_payoff_vanilla, plot_payoff_straddle, plot_payoff_strangle
 
long_straddle = Strategy(k=100, t=.2, style="straddle", way="long")
short_straddle = Strategy(k=100, t=.2, style="straddle", qty=-1, way="short")
long_strangle = Strategy(k=80, k2=120, t=.5, style="strangle",qty=1)
short_strangle= Strategy(k=80, k2=120, t=.5, style="strangle",qty=-1, way="short")

bsm = BlackScholesModel(spot=100, r=0.05, sigma=0.3)
long_straddle.pricer(model=bsm)
short_straddle.pricer(model=bsm)
long_strangle.pricer(model=bsm)
short_strangle.pricer(model=bsm)


plot_payoff_straddle(long_straddle)
plot_payoff_straddle(short_straddle)

plot_payoff_strangle(long_strangle)
plot_payoff_strangle(short_strangle)
```
